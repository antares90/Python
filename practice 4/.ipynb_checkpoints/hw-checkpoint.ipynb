```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from sklearn import preprocessing

df = pd.read_csv("train.csv")
# print(df)
print(df.shape)
# 1. Удаление бесполезных колонок
df = df.drop(['PassengerId', 'SibSp', 'Parch', 'Cabin', 'Embarked', 'Ticket','Name'], axis=1)
print(df.shape)
df.head()
```

```python
# поиск NaN
for i in df.columns:
    print(i,': ',np.sum(df[i].isna()))
```

```python
df.head()
```

```python
# избавляемся от NaN
df_age = df['Age'].fillna(df['Age'].mean())
    
# df = df.drop(['Age'], axis=1)
df['Age'] = df_age
df.head()
```

```python
# 2. Примените one-hot encoding к полю Sex.
print(df.nunique())
df_sex = pd.get_dummies(df['Sex'])

for i in df_sex.columns:
    df[i] = df_sex[i]

    
df = df.drop(['Sex'], axis=1)
df.head()
```

```python
#3. стандартизация возраста и стоимости билетов
# standardized_X = preprocessing.scale(df['Age'])
df['Age'] = (df['Age'] - df['Age'].mean()) / df['Age'].std()
df['Fare'] = (df['Fare'] - df['Fare'].mean()) / df['Fare'].std()
df.head()
```
