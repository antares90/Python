```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('train.csv') 

#формирование векторов для графика
dis = df.sort_values(by='dis')['dis']
medv = df['medv']

#построение графика
plt.grid(True)
plt.xlabel('dis')
plt.ylabel('medv')
plt.title('График зависимости значений medv от значений dis')
plt.plot(dis, medv)
```

```python
#формирование данных для домов, чья цена выше средней цены всех домов
print('среднее значение medv = ', medv.mean())
df_rich = df[df['medv']>medv.mean()].sort_values(by='medv').head(10)
# сохранение данных в файл
df_rich.to_csv('rich_boston.csv')
df_rich
```
