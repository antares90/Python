## Работа с текстом

```python
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2*np.pi*x)
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('Time in seconds')
plt.ylabel('Signal')
plt.title('Plot of signal distribution over time')
plt.grid(True)
```

## Несколько графиков

```python
def f1(t):
    return np.exp(-t) * np.cos(t)

def f2(t):
    return np.cos(t)

t = np.arange(0.0, 5.0, 0.1)

plt.subplot(121)
plt.plot(t, f1(t), 'bo')

plt.subplot(122)
plt.plot(t, f2(t), 'r--')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
