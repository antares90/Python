```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

#чтение файла
df = pd.read_csv('hw2.csv') 
df.head()
X = np.array(df['X'])
Y = np.array(df['Y'])
```

```python
#создание модели
model = LinearRegression()
model
model.fit(X.reshape(-1, 1), Y.reshape(-1, 1))
```

```python
#получение предсказаных результатов
Y_pred = model.predict(X.reshape(-1, 1))
```

```python
plt.plot(X,Y, 'o')
plt.plot(X, Y_pred)
```

```python
W = model.coef_
b = model.intercept_
print('W = ', W)
print('b = ', b)
```
