```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

#чтение файла
df = pd.read_csv('train.csv') 
x = np.array(df['dis'])
y = np.array(df['medv'])
df
```

```python
#Задание гиперпараметров
alpha = 0.000001
n = 500
```

```python
W = np.array([0])
b = np.array([0])
```

```python
def net_input(W, X, b):    
    return np.dot(W, X) + b
```

```python
def cost_function_derivative_w(Y_hat, Y, X):
    return 2 * np.sum((Y_hat - Y) * X) / Y.shape[0]
```

```python
def cost_function_derivative_b(Y_hat, Y):
    return 2 * np.sum(Y_hat - Y) / Y.shape[0]
```

```python
for i in range(0, n):
    #     y_hat = np.array([net_input(W[i], [x_single], b) for x_single in x.T[i])   
    y_hat = np.array([net_input(W, [x_single], b) for x_single in x])
    W = W - alpha * cost_function_derivative_w(y_hat, y.reshape(-1, 1), x)
    b = b - alpha * cost_function_derivative_b(y_hat, y.reshape(-1, 1))
print(W)
print(b)
y_hat
```

```python
plt.plot(x, y, 'o')
plt.plot(np.arange(0, 10), np.arange(0, 10) * W + b, 'r--')
plt.show()
```
